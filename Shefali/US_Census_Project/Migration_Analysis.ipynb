```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Creating Data Frames for Each csv files

```python
age_data = pd.read_csv('Age_Data.csv')
age_data_df = pd.DataFrame(age_data)
age_data_df = age_data_df[~age_data['Age'].isin(['1 to 4 years', '5 to 17 years', '75 years and over'])]
age_data_df.head()
```

```python
education_data = pd.read_csv('Education_Data.csv')
education_data_df = pd.DataFrame(education_data)
education_data_df['Education Attainment'] = education_data_df['Education Attainment'].str.strip()
education_data_df.head()

# filter to year 2013
#education_data_df = education_data_df[education_data_df['Year'] == 2014]
```

```python
gender_data = pd.read_csv('Gender_Data.csv')
gender_data_df = pd.DataFrame(gender_data)
gender_data_df['Sex'] = gender_data_df['Sex'].str.strip()
gender_data_df.head()
```

```python
race_data = pd.read_csv('Race_Data.csv')
race_data_df = pd.DataFrame(race_data)
race_data_df['Race'] = race_data_df['Race'].str.strip()
race_data_df = race_data_df[~race_data_df['Race'].isin(['White alone. not Hispanic or Latino', 'Some other race'])]

race_data_df
```

#### Plotting Migration of Each Age Group

```python
# Plot for "Moved within same county"
plt.figure(figsize=(12, 8))
sns.lineplot(data=age_data_df, x='Year', y='Moved within same county', hue='Age', marker='o')
plt.title('Moved within Same County (2012-2022)')
plt.ylabel('% Population')
plt.legend(title='Age Group', loc='upper right')
plt.show()
```

```python
# Plot for "Moved from different county, same state"
plt.figure(figsize=(12, 8))
sns.lineplot(data=age_data_df, x='Year', y='Moved from different county. same state', hue='Age', marker='o')
plt.title('Moved from Different County, Same State (2012-2022)')
plt.ylabel('% Population')
plt.legend(title='Age Group', loc='lower right')
plt.show()
```

```python
# Plot for "Moved from different state"
plt.figure(figsize=(12, 8))
sns.lineplot(data=age_data_df, x='Year', y='Moved from different state', hue='Age', marker='o')
plt.title('Moved from Different State (2012-2022)')
plt.ylabel('% Population')
plt.legend(title='Age Group', loc='lower right')
plt.show()
```

```python
# Plot for "Moved from abroad"
plt.figure(figsize=(12, 8))
sns.lineplot(data=age_data_df, x='Year', y='Moved from abroad', hue='Age', marker='o')
plt.title('Moved from Abroad (2012-2022)')
plt.ylabel('% Population')
plt.legend(title='Age Group', loc='lower right')
plt.show()
```
